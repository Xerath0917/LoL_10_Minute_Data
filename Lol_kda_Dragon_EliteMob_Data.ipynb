```python
#KDA_Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

lol_kda_data = pd.read_csv('high_diamond_ranked_10min.csv')

#BlueTeamKdaData
blue_kills = list(lol_kda_data['blueKills'])
blue_death = list(lol_kda_data['blueDeaths'])
blue_assist = list(lol_kda_data['blueAssists'])

blue_kda_list = []
def bluekda(k, d, a):
    for kl, dl, al in zip(k, d, a):
        if dl == 0:
            dl = 1
            blue_kda = (kl + al) / dl
            blue_kda_list.append(blue_kda)
        else:
            blue_kda = (kl + al) / dl
            blue_kda_list.append(blue_kda)

bluekda(blue_kills, blue_death, blue_assist)

blue_kda_list = pd.DataFrame(blue_kda_list)
blue_kda_list.columns = ['Blue_KDA']

x1 = blue_kda_list.index
y1 = blue_kda_list['Blue_KDA']

plt.plot(x1, y1 ,color = '#50C8FF')

#RedTeamKdaData
red_kills = list(lol_kda_data['redKills'])
red_death = list(lol_kda_data['redDeaths'])
red_assist = list(lol_kda_data['redAssists'])

red_kda_list = []
def redkda(k, d, a):
    for kl, dl, al in zip(k, d, a):
        if dl == 0:
            dl = 1
            red_kda = (kl + al) / dl
            red_kda_list.append(red_kda)
        else:
            red_kda = (kl + al) / dl
            red_kda_list.append(red_kda)

redkda(red_kills, red_death, red_assist)

red_kda_list = pd.DataFrame(red_kda_list)
red_kda_list.columns = ['Red_KDA']

x2 = red_kda_list.index 
y2 =red_kda_list['Red_KDA']
plt.plot(x2, y2 ,color = '#FF98A3')
plt.title('10 Minute Team Kda')
plt.xlabel('Game_Id')
plt.ylabel('Kda')

plt.show()
```

```python
#Team_Dragons_Data

import pandas as pd
import matplotlib.pyplot as plt
lol_data = pd.read_csv('high_diamond_ranked_10min.csv')

dragons = pd.DataFrame(lol_data['blueDragons'])
dragons['blueWins'] = lol_data['blueWins']
dragons_l = list(dragons['blueDragons'])
dragons_t = list(dragons['blueWins'])

dragons_tt = []

for f_l, f_t in zip(dragons_l, dragons_t):
    if f_l == 1:
        if f_t ==1:
            dragons_tt.append(1)
        else:
            dragons_tt.append(0)
    else:
        dragons_tt.append(0)
dragons_tt = pd.DataFrame(dragons_tt)
dragons['test'] = dragons_tt

labels = 'redwins' , 'bluewins'
sizes = [dragons.test[dragons['test'] == 1].count(), dragons.test[dragons['test'] == 0].count()]
colors = ['#FF98A3', '#50C8FF']
explode = (0,0.1)
fig1,ax1 = plt.subplots(figsize = (5, 5))
ax1.pie(sizes, explode = explode, labels= labels, autopct='%1.1f%%', shadow = True, startangle = 90, colors=colors)
ax1.axis('equal')
plt.title('Team Dragons')
plt.show
```
